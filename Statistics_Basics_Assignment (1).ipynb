# Statistics Basics Assignment

1. What is statistics, and why is it important??

 Statistics is a branch of mathematics that deals with collecting, analyzing, interpreting, presenting, and organizing data. It helps in making informed decisions by providing insights from data. It is important because it supports scientific research, business decision-making, government policy development, and much more.

2. What are the two main types of statistics?
  - Descriptive Statistics
  - Inferential Statistics

3. What are descriptive statistics?

 Descriptive statistics summarize or describe the main features of a dataset through measures such as mean, median, mode, variance, and standard deviation.

4. What is inferential statistics?

 Inferential statistics use a sample of data to make inferences or generalizations about a population using probability theory.

5. What is sampling in statistics?

 Sampling is the process of selecting a subset of individuals from a population to estimate characteristics of the whole population.

6. What are the different types of sampling methods?

 - Simple Random Sampling
 - Systematic Sampling
 - Stratified Sampling
 - Cluster Sampling
 - Convenience Sampling

7. What is the difference between random and non-random sampling?

 Random sampling gives all individuals equal chance of selection; non-random sampling does not follow this principle, often leading to bias.

8. Define and give examples of qualitative and quantitative data.

 - Qualitative data: Categorical data (e.g., gender, color)
 - Quantitative data: Numerical data (e.g., height, weight)

9. What are the different types of data in statistics?

 - Nominal
 - Ordinal
 - Interval
 - Ratio

10. Explain nominal, ordinal, interval, and ratio levels of measurement.

 - Nominal: Categories without order (e.g., colors)
 - Ordinal: Ordered categories (e.g., rankings)
 - Interval: Numeric with equal intervals, no true zero (e.g., temperature)
 - Ratio: Numeric with true zero (e.g., weight)

11. What is the measure of central tendency?

 It refers to measures that define the center of a dataset—mean, median, and mode.

12. Define mean, median, and mode.

 - Mean: Average of values
 - Median: Middle value
 - Mode: Most frequent value

13. What is the significance of the measure of central tendency?

It helps in understanding the distribution and central point of the dataset.

14. What is variance, and how is it calculated?

 Variance measures the spread of data points from the mean. It is the average of squared deviations from the mean.

15. What is standard deviation, and why is it important?

 Standard deviation is the square root of variance. It indicates how much the data deviates from the mean, aiding in risk and variability assessment.

16. Define and explain the term range in statistics.

 Range = Max value - Min value. It indicates the spread of values in the dataset.

17. What is the difference between variance and standard deviation?

 Variance is in squared units; standard deviation is in original units. Standard deviation is more interpretable.

18. What is skewness in a dataset?

 Skewness indicates asymmetry in the distribution of data.

19. What does it mean if a dataset is positively or negatively skewed?

 - Positive skew: Tail on the right side
 - Negative skew: Tail on the left side

20. Define and explain kurtosis.

 Kurtosis measures the "tailedness" of a distribution—whether data has heavy or light tails compared to a normal distribution.

21. What is the purpose of covariance?

 Covariance measures the direction of the linear relationship between two variables.

22. What does correlation measure in statistics?

 Correlation quantifies the strength and direction of a linear relationship between two variables.

23. What is the difference between covariance and correlation?

 Covariance indicates direction; correlation standardizes the relationship between -1 and 1.

24. What are some real-world applications of statistics?
- Business decision-making
- Market analysis
- Health research
- Government policy
- Quality control
- Education

```python
# 4. Implement sampling techniques like random sampling and stratified sampling.

import pandas as pd
from sklearn.model_selection import train_test_split

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Hannah'],
    'Gender': ['F', 'M', 'M', 'M', 'F', 'M', 'F', 'F'],
    'Score': [85, 90, 78, 88, 92, 76, 81, 95]
}

df = pd.DataFrame(data)

random_sample = df.sample(frac=0.5, random_state=1)
print("Random Sample:\n", random_sample)
stratified_sample, _ = train_test_split(df,
                                        test_size=0.5,
                                        stratify=df['Gender'],
                                        random_state=1)
print("\nStratified Sample:\n", stratified_sample)
```

```python
# 7. Calculate skewness and kurtosis of a dataset using Python libraries.

import numpy as np
import pandas as pd
from scipy.stats import skew, kurtosis

data = np.random.exponential(scale=2.0, size=1000)

series = pd.Series(data)
skewness = skew(series)
kurt = kurtosis(series)

print("Skewness:", skewness)
print("Kurtosis:", kurt)
```

```python
# 8. Generate a dataset and demonstrate positive and negative skewness

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew


pos_skewed = np.random.exponential(scale=2.0, size=1000)


neg_skewed = -np.random.exponential(scale=2.0, size=1000)


fig, axs = plt.subplots(1, 2, figsize=(14, 5))


sns.histplot(pos_skewed, bins=30, kde=True, ax=axs[0], color='orange')
axs[0].set_title(f"Positive Skew\nSkewness = {skew(pos_skewed):.2f}")
axs[0].set_xlabel("Value")

sns.histplot(neg_skewed, bins=30, kde=True, ax=axs[1], color='green')
axs[1].set_title(f"Negative Skew\nSkewness = {skew(neg_skewed):.2f}")
axs[1].set_xlabel("Value")

plt.tight_layout()
plt.show()

```

```python
# 9. Write a Python script to calculate covariance between two datasets.

import numpy as np
import pandas as pd

x = [10, 20, 30, 40, 50]
y = [15, 25, 35, 45, 60]

cov_matrix = np.cov(x, y, bias=False)  # Unbiased estimator
print("Covariance using NumPy:", cov_matrix[0, 1])
df = pd.DataFrame({'X': x, 'Y': y})
cov_value = df.cov().iloc[0, 1]
print("Covariance using Pandas:", cov_value)
```

```python
# 10. Write a Python script to calculate the correlation coefficient between two datasets.
import numpy as np
import pandas as pd

x = [10, 20, 30, 40, 50]
y = [15, 25, 35, 45, 60]


correlation_matrix = np.corrcoef(x, y)
print("Correlation coefficient using NumPy:", correlation_matrix[0, 1])

df = pd.DataFrame({'X': x, 'Y': y})
corr_value = df.corr().iloc[0, 1]
print("Correlation coefficient using Pandas:", corr_value)
```

```python
# 12. Implement and compare simple random sampling and systematic sampling.

import pandas as pd
import numpy as np

data = pd.DataFrame({'ID': range(1, 101)})

random_sample = data.sample(n=10, random_state=1)
print("Simple Random Sample:\n", random_sample)
k = 10
start = np.random.randint(0, k)
systematic_sample = data.iloc[start::k]
print("\nSystematic Sample (every 10th starting from {}):\n".format(start), systematic_sample)
```

```python
# 14. Simulate data using Python and calculate its central tendency and dispersion.

import numpy as np
import pandas as pd
import statistics as stats

data = np.random.normal(loc=50, scale=10, size=1000)
series = pd.Series(data)

mean_val = series.mean()
median_val = series.median()
mode_val = series.mode()[0]

# Dispersion
range_val = series.max() - series.min()
variance_val = series.var()
std_dev = series.std()
iqr_val = series.quantile(0.75) - series.quantile(0.25)


print(f"Mean: {mean_val:.2f}")
print(f"Median: {median_val:.2f}")
print(f"Mode: {mode_val:.2f}")
print(f"Range: {range_val:.2f}")
print(f"Variance: {variance_val:.2f}")
print(f"Standard Deviation: {std_dev:.2f}")
print(f"IQR: {iqr_val:.2f}")

```

```python
# 17. Calculate the interquartile range (IQR) of a dataset.

import numpy as np
import pandas as pd

data = np.random.normal(loc=50, scale=15, size=1000)
series = pd.Series(data)

Q1 = series.quantile(0.25)
Q3 = series.quantile(0.75)
IQR = Q3 - Q1

print(f"Q1 (25th percentile): {Q1:.2f}")
print(f"Q3 (75th percentile): {Q3:.2f}")
print(f"IQR: {IQR:.2f}")
```

```python
# 18. Implement Z-score normalization and explain its significance.

import numpy as np
import pandas as pd

# Simulate a dataset
data = np.random.normal(100, 20, 10)
df = pd.DataFrame({'Original': data})

# Z-score normalization
mean = df['Original'].mean()
std = df['Original'].std()
df['Z_score'] = (df['Original'] - mean) / std

print(df)
```

```python
# 19. Compare two datasets using their standard deviations.

import numpy as np
import pandas as pd

dataset_A = np.random.normal(50, 5, 1000)
dataset_B = np.random.normal(50, 20, 1000)


df = pd.DataFrame({'Dataset A': dataset_A, 'Dataset B': dataset_B})

std_A = df['Dataset A'].std()
std_B = df['Dataset B'].std()

print(f"Standard Deviation of Dataset A: {std_A:.2f}")
print(f"Standard Deviation of Dataset B: {std_B:.2f}")


if std_A < std_B:
    print("Dataset A is more consistent (less variable) than Dataset B.")
else:
    print("Dataset B is more consistent (less variable) than Dataset A.")
```

```python
# 20. Write a Python program to visualize covariance using a heatmap.

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


np.random.seed(42)
data = {
    'Height': np.random.normal(160, 10, 100),
    'Weight': np.random.normal(65, 12, 100),
    'Age': np.random.normal(35, 8, 100)
}

df = pd.DataFrame(data)
cov_matrix = df.cov()
plt.figure(figsize=(6, 4))
sns.heatmap(cov_matrix, annot=True, cmap="YlGnBu", fmt=".2f")
plt.title("Covariance Heatmap")
plt.show()
```

```python
# 21. Use seaborn to create a correlation matrix for a dataset.

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(0)
data = {
    'Math': np.random.normal(70, 10, 100),
    'Science': np.random.normal(65, 15, 100),
    'English': np.random.normal(75, 12, 100),
    'Attendance': np.random.normal(90, 5, 100)
}

df = pd.DataFrame(data)
corr_matrix = df.corr()

plt.figure(figsize=(6, 4))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
# 23. Visualize skewness and kurtosis using Python libraries like matplotlib or seaborn

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis

data = np.random.exponential(scale=2, size=1000)
series = pd.Series(data)

sk = skew(series)
kt = kurtosis(series)
plt.figure(figsize=(8, 4))
sns.histplot(series, bins=30, kde=True, color='skyblue')
plt.title(f"Distribution with Skewness = {sk:.2f}, Kurtosis = {kt:.2f}")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
