# Statistics Basics Assignment

1. What is statistics, and why is it important??

 Statistics is a branch of mathematics that deals with collecting, analyzing, interpreting, presenting, and organizing data. It helps in making informed decisions by providing insights from data. It is important because it supports scientific research, business decision-making, government policy development, and much more.

2. What are the two main types of statistics?
  - Descriptive Statistics
  - Inferential Statistics

3. What are descriptive statistics?

 Descriptive statistics summarize or describe the main features of a dataset through measures such as mean, median, mode, variance, and standard deviation.

4. What is inferential statistics?

 Inferential statistics use a sample of data to make inferences or generalizations about a population using probability theory.

5. What is sampling in statistics?

 Sampling is the process of selecting a subset of individuals from a population to estimate characteristics of the whole population.

6. What are the different types of sampling methods?

 - Simple Random Sampling
 - Systematic Sampling
 - Stratified Sampling
 - Cluster Sampling
 - Convenience Sampling

7. What is the difference between random and non-random sampling?

 Random sampling gives all individuals equal chance of selection; non-random sampling does not follow this principle, often leading to bias.

8. Define and give examples of qualitative and quantitative data.

 - Qualitative data: Categorical data (e.g., gender, color)
 - Quantitative data: Numerical data (e.g., height, weight)

9. What are the different types of data in statistics?

 - Nominal
 - Ordinal
 - Interval
 - Ratio

10. Explain nominal, ordinal, interval, and ratio levels of measurement.

 - Nominal: Categories without order (e.g., colors)
 - Ordinal: Ordered categories (e.g., rankings)
 - Interval: Numeric with equal intervals, no true zero (e.g., temperature)
 - Ratio: Numeric with true zero (e.g., weight)

11. What is the measure of central tendency?

 It refers to measures that define the center of a dataset—mean, median, and mode.

12. Define mean, median, and mode.

 - Mean: Average of values
 - Median: Middle value
 - Mode: Most frequent value

13. What is the significance of the measure of central tendency?

It helps in understanding the distribution and central point of the dataset.

14. What is variance, and how is it calculated?

 Variance measures the spread of data points from the mean. It is the average of squared deviations from the mean.

15. What is standard deviation, and why is it important?

 Standard deviation is the square root of variance. It indicates how much the data deviates from the mean, aiding in risk and variability assessment.

16. Define and explain the term range in statistics.

 Range = Max value - Min value. It indicates the spread of values in the dataset.

17. What is the difference between variance and standard deviation?

 Variance is in squared units; standard deviation is in original units. Standard deviation is more interpretable.

18. What is skewness in a dataset?

 Skewness indicates asymmetry in the distribution of data.

19. What does it mean if a dataset is positively or negatively skewed?

 - Positive skew: Tail on the right side
 - Negative skew: Tail on the left side

20. Define and explain kurtosis.

 Kurtosis measures the "tailedness" of a distribution—whether data has heavy or light tails compared to a normal distribution.

21. What is the purpose of covariance?

 Covariance measures the direction of the linear relationship between two variables.

22. What does correlation measure in statistics?

 Correlation quantifies the strength and direction of a linear relationship between two variables.

23. What is the difference between covariance and correlation?

 Covariance indicates direction; correlation standardizes the relationship between -1 and 1.

24. What are some real-world applications of statistics?
- Business decision-making
- Market analysis
- Health research
- Government policy
- Quality control
- Education
